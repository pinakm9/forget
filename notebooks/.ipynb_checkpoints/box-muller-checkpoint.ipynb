```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Generate uniform random numbers U1 and U2
num_samples = 500
U1 = np.random.uniform(0, 1, num_samples)
U2 = np.random.uniform(0, 1, num_samples)

# Step 2: Convert to polar coordinates (R, Î¸)
R = np.sqrt(-2 * np.log(U1))
theta = 2 * np.pi * U2

# Step 3: Convert back to Cartesian coordinates (Z0, Z1) -> Normal Distribution
Z0 = R * np.cos(theta)
Z1 = R * np.sin(theta)

# Create figure
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Initialize scatter plots
scatter1 = axs[0].scatter([], [])
scatter2 = axs[1].scatter([], [])
scatter3 = axs[2].scatter([], [])

# Titles
axs[0].set_title("Step 1: Uniform Distribution")
axs[1].set_title("Step 2: Polar Coordinates")
axs[2].set_title("Step 3: Gaussian Distribution")

# Limits
axs[0].set_xlim(0, 1)
axs[0].set_ylim(0, 1)

axs[1].set_xlim(-1, 1)
axs[1].set_ylim(-1, 1)

axs[2].set_xlim(-4, 4)
axs[2].set_ylim(-4, 4)

# Labels
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# Animation update function
def update(frame):
    scatter1.set_offsets(np.c_[U1[:frame], U2[:frame]])
    scatter2.set_offsets(np.c_[np.cos(theta[:frame]), np.sin(theta[:frame])])
    scatter3.set_offsets(np.c_[Z0[:frame], Z1[:frame]])
    return scatter1, scatter2, scatter3

# Create animation
ani = animation.FuncAnimation(fig, update, frames=num_samples, interval=20, blit=False)

# Save the animation as an MP4 file
ani.save("../data/MNIST/box_muller_animation.mp4", writer="ffmpeg", fps=30)
print("Animation saved as 'box_muller_animation.mp4' ðŸŽ¥")
```
